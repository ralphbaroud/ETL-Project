```python
# Dependencies
import pandas as pd
```

```python
file_2015 = "worldhappy2015.csv"
file_2016 = "worldhappy2016.csv"
file_2017 = "worldhappy2017.csv"
file_2018 = "worldhappy2018.csv"
file_2019e = "worldhappy2019extra.csv"
```

```python

file_2015_df = pd.read_csv(file_2015, encoding="ISO-8859-1")
file_2015_df.head()
```

```python
file_2016_df = pd.read_csv(file_2016, encoding="ISO-8859-1")
cols = [0,1,2,3]
file_2016_df1 = file_2016_df[file_2016_df.columns[cols]]
file_2016_df1.head()
```

```python
file_2017_df = pd.read_csv(file_2017, encoding="ISO-8859-1")
file_2017_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_2017_df = file_2017_df.rename(columns={"Happiness.Rank":"Happiness Rank", "Happiness.Score":"Happiness Score"})
renamed_2017_df.head()
```

```python
file_2018_df = pd.read_csv(file_2018, encoding="ISO-8859-1")
file_2018_df.head()
```

```python
# Renaming the columns 
renamed_2018_df = file_2018_df.rename(columns={"Country or region":"Country", "ï»¿Overall rank":"Happiness Rank", "Score":"Happiness Score"})
renamed_2018_df.head()
```

```python
# Reorganizing the columns using double brackets
organized_2018_df = renamed_2018_df[["Country","Happiness Rank","Happiness Score","Year"]]
organized_2018_df.head()
```

```python
file_2019e_df = pd.read_csv(file_2019e, encoding="ISO-8859-1")
file_2019e_df.head()

```

```python
renamed_2019e_df = file_2019e_df.rename(columns={"Countries":"Country", "World Happiness Ranking":"Happiness Rank", "World Happiness Index":"Happiness Score"})
renamed_2019e_df.head()


```

```python
# sort the last of the datasets (renamed_2019e_df) by Happiness Rank
sorted_2019e_df = renamed_2019e_df.sort_values("Happiness Rank")
sorted_2019e_df.head()
```

```python
# concat all 5 files
concat_happy_df = pd.concat([file_2015_df, file_2016_df1, renamed_2017_df, organized_2018_df, sorted_2019e_df],axis=0,sort=False).reset_index()
concat_happy_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
concat_happy_df.to_csv("output_happy.csv", index=False, header=True)
```
